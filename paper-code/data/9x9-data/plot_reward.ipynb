```python
import numpy as np
import os
import matplotlib.pyplot as plt

def window(size):
    return np.ones(size)/float(size)

avg_reward_no_shield = np.loadtxt(os.getcwd()+
    '/9x9_illustrative_0_avg_reward.data')
ep_len_no_shield = np.loadtxt(os.getcwd()+
    '/9x9_illustrative_0_episodelen.data')

avg_reward_shield_1_action = np.loadtxt(os.getcwd()+
    '/9x9_illustrative_1_avg_reward.data')
ep_len_shield_1_action = np.loadtxt(os.getcwd()+
    '/9x9_illustrative_1_episodelen.data')

avg_reward_shield_1_action_neg_reward = np.loadtxt(os.getcwd()+
    '/9x9_illustrative_1_neg_reward_avg_reward.data')
ep_len_shield_1_action_neg_reward = np.loadtxt(os.getcwd()+
    '/9x9_illustrative_1_neg_reward_episodelen.data')

avg_reward_shield_3_action = np.loadtxt(os.getcwd()+
    '/9x9_illustrative_3_avg_reward.data')
ep_len_shield_3_action = np.loadtxt(os.getcwd()+
    '/9x9_illustrative_3_episodelen.data')

avg_reward_shield_3_action_neg_reward = np.loadtxt(os.getcwd()+
    '/9x9_illustrative_3_neg_reward_avg_reward.data')
ep_len_shield_3_action_neg_reward = np.loadtxt(os.getcwd()+
    '/9x9_illustrative_3_neg_reward_episodelen.data')

t = 10
plt.plot(np.convolve(avg_reward_no_shield,window(t),'same'), label='no shield')
plt.plot(np.convolve(avg_reward_shield_1_action,window(t),'same'), label='shield-1 action')
plt.plot(np.convolve(avg_reward_shield_1_action_neg_reward,window(t),'same'), label='shield-1 action-neg reward')
plt.plot(np.convolve(avg_reward_shield_3_action,window(t),'same'), label='shield-3 action')
plt.plot(np.convolve(avg_reward_shield_3_action_neg_reward,window(t),'same'), label='shield-3 action-neg reward')

plt.legend()

plt.show()


plt.plot(np.convolve(ep_len_no_shield,window(t),'same'), label='no shield')
plt.plot(np.convolve(ep_len_shield_1_action,window(t),'same')[:45], label='shield-1 action')
plt.plot(np.convolve(ep_len_shield_1_action_neg_reward,window(t),'same'), label='shield-1 action-neg reward')
plt.plot(np.convolve(ep_len_shield_3_action,window(t),'same'), label='shield-3 action')
plt.plot(np.convolve(ep_len_shield_3_action_neg_reward,window(t),'same'), label='shield-3 action-neg reward')

plt.legend()

plt.show()
```
